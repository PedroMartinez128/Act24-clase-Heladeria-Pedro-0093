```python
class Heladeria:
    def __init__(self, nombre, sabores, temperatura, ubicacion):
        # Atributos (Datos)
        self.nombre = nombre
        self.saboresDisponibles = sabores
        self.temperaturaFreezer = temperatura
        self.ubicacion = ubicacion

    # Funciones (Acciones)
    def preparar_helado(self):
        print(f"\n👨‍🍳 [PREPARANDO] En {self.nombre}...")
        sabor = input(f"Sabor a preparar ({self.saboresDisponibles}): ")
        print(f"🌀 Mezclando ingredientes a {self.temperaturaFreezer}...")
        print(f"✅ ¡Tu helado de {sabor} está listo!")

    def servir_cono(self):
        print(f"🍦 [SERVICIO] Poniendo la bola en un cono crujiente...")
        print(f"📍 ¡Disfrútalo en nuestra sucursal de {self.ubicacion}!")

    def cobrar_pedido(self):
        monto = input("💰 [CAJA] Ingrese el monto a cobrar: $")
        print(f"💳 Cobro de ${monto} procesado con éxito en {self.nombre}. ¡Gracias! 👋")

# --- PASO 1: Registro de la Heladería 1 ---
print("🌟 BIENVENIDO AL REGISTRO DE HELADERÍAS 🌟")
print("\n--- DATOS DE LA HELADERÍA 1 ---")
h1_nom = input("Nombre del local: ")
h1_sab = input("Sabores (ej: Fresa, Limón): ")
h1_tem = input("Temperatura (ej: -18°C): ")
h1_ubi = input("Dirección: ")

heladeria1 = Heladeria(h1_nom, h1_sab, h1_tem, h1_ubi)

# --- PASO 2: Registro de la Heladería 2 ---
print("\n--- DATOS DE LA HELADERÍA 2 ---")
h2_nom = input("Nombre del local: ")
h2_sab = input("Sabores (ej: Pistacho, Menta): ")
h2_tem = input("Temperatura (ej: -15°C): ")
h2_ubi = input("Dirección: ")

heladeria2 = Heladeria(h2_nom, h2_sab, h2_tem, h2_ubi)

# --- PASO 3: Ejecución para Heladería 1 ---
print(f"\n🚀 ABRIENDO LAS PUERTAS DE: {heladeria1.nombre} 🚀")
heladeria1.preparar_helado()
heladeria1.servir_cono()
heladeria1.cobrar_pedido()

# --- PASO 4: Ejecución para Heladería 2 ---
print(f"\n🚀 ABRIENDO LAS PUERTAS DE: {heladeria2.nombre} 🚀")
heladeria2.preparar_helado()
heladeria2.servir_cono()
heladeria2.cobrar_pedido()

print("\n✨ ¡Programa realizado por Pedro Martinez NL = 34. ✨")
```
